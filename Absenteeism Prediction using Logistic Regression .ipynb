# Importing Relevant Libraries

```python
import pandas as pd
import numpy as np
```

# Load the Data

```python
raw_csv_data = pd.read_csv("Absenteeism_data.csv")
```

# Data Processing

```python
#creating a copy of the data to ensure data integrity and prevent any loss or corruption
df = raw_csv_data.copy()
df.head(10)
```

```python
df.describe()
```

```python
#dropping the ID column
df = df.drop(['ID'], axis = 1)
df.head()
```

```python
reason_columns = pd.get_dummies(df['Reason for Absence'])

reason_columns['check'] = reason_columns.sum(axis=1)
reason_columns = reason_columns.drop(['check'], axis = 1)


reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first = True)
```

```python
#dropping the column 'Reason for Absence'
df = df.drop(['Reason for Absence'], axis = 1)
df.head()
```

```python
#replacing the 'Reason for Absence' column with 4 columns 
reason_type_1 = reason_columns.loc[:, 1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:, 15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:, 18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:, 22:].max(axis=1)

df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis = 1)
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']

df.columns = column_names
```

```python
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 
                          'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']

df = df[column_names_reordered]

df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format = '%d/%m/%Y')
```

```python
list_months = []
```

```python
for i in range(df.shape[0]):
    list_months.append(df['Date'][i].month)
    
# list_months
# len(list_months)
df['Month Value'] = list_months
# df_reason_mod.head(20)
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()

df['Day of the Week'] = df['Date'].apply(date_to_weekday)
```

```python
df = df.drop(['Date'], axis = 1)
# df_reason_mod.head()
# df_reason_mod.columns.values
column_names_upd = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value', 'Day of the Week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
       'Pets', 'Absenteeism Time in Hours']

df.head()
```

```python
#Converting the values in the education column into binary
df['Education'] = df['Education'].map({1:0, 2:1, 3:1, 4:1})
```

```python
data_preprocessed = df.copy()
data_preprocessed
```

# Create the targets

```python
data_preprocessed['Absenteeism Time in Hours'].median()
#Here we are using median instead of mean beacause the median is resistant to extreme values or outliers
```

```python
targets = np.where(data_preprocessed['Absenteeism Time in Hours'] > 
                   data_preprocessed['Absenteeism Time in Hours'].median(), 1, 0)
```

```python
targets
```

```python
# create a Series in the original data frame that will contain the targets for the regression analysis
data_preprocessed['Excessive Absenteeism'] = targets
data_preprocessed.head()
```

# Target Analysis

```python
# check if dataset is balanced (what % of targets are 1s)
targets.sum() / targets.shape[0]
```

```python
data_with_targets = data_preprocessed.drop(['Absenteeism Time in Hours',],axis=1)

#Here we can drop columns such as 'Day of the Week','Daily Work Load Average','Distance to Work' also
#after initially testing we will find that these factors have very affect in determining absenteeism
#removing these columns will increase the accuracy by a few percentages
```

```python
data_with_targets.head()
```

# Selecting inputs for regression

```python
data_with_targets.shape
```

```python
data_with_targets.iloc[:,:14]
```

```python
data_with_targets.iloc[:,:-1]
```

```python
unscaled_inputs = data_with_targets.iloc[:,:-1]
#this variable will contain the inputs(everything except the target variable)
```

# Standardize the Data

```python
from sklearn.preprocessing import StandardScaler
absenteeism_scaler = StandardScaler()
```

```python
absenteeism_scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = absenteeism_scaler.transform(unscaled_inputs)
```

```python
scaled_inputs
```

```python
scaled_inputs.shape
```

#   Split the data into train data and test data

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split(scaled_inputs, targets)
```

```python
# declare 4 variables for the split
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets, #train_size = 0.75, 
                                                                            test_size = 0.25, random_state = 20)
```

```python
print (x_train.shape, y_train.shape)
print (x_test.shape, y_test.shape)
```

# Logistic regression with sklearn

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
#this will asses the training accuracy of the model
reg.score(x_train,y_train)
```

# Manually checking the accuracy

```python
model_outputs = reg.predict(x_train)
model_outputs
```

# Finding the intercept and coefficients

```python
#get the intercept of our model
reg.intercept_
```

```python
#get the coefficients of our model
reg.coef_
```

```python
unscaled_inputs.columns.values
```

```python
feature_name = unscaled_inputs.columns.values
```

```python
summary_table = pd.DataFrame (columns=['Feature name'], data = feature_name)

summary_table['Coefficient'] = np.transpose(reg.coef_)
summary_table
```

```python
#move the intercept to the top of the summary table
summary_table.index = summary_table.index + 1

summary_table.loc[0] = ['Intercept', reg.intercept_[0]]

summary_table = summary_table.sort_index()
summary_table
```

# Interpreting the coefficients

```python
summary_table['Odds_ratio'] = np.exp(summary_table.Coefficient)
```

```python
summary_table
```

```python
summary_table.sort_values('Odds_ratio', ascending=False)
```

# Testing the model

```python
reg.score(x_test,y_test)
```
